```python
import os
import random
import csv
from PIL import Image
```

```python
PATH = "/kaggle/input/vietnamese-license-plate-ocr/cropped"
PATH = "/kaggle/input/vietnamese-license-plate-ocr/cropped"
OUT_IMG = "/kaggle/working/imgs"
OUT_LABEL = "/kaggle/working/labels"

os.makedirs(f"{OUT_IMG}/train", exist_ok=True)
os.makedirs(f"{OUT_IMG}/val", exist_ok=True)
os.makedirs(OUT_LABEL, exist_ok=True)

paths = [f for f in os.listdir(PATH) if f.lower().endswith((".jpg", ".png", ".jpeg"))]

random.seed(10)
random.shuffle(paths)

train_len = int(len(paths) * 0.8)
train_imgs = paths[:train_len]
val_imgs   = paths[train_len:]

for img_name in train_imgs:
    img = Image.open(os.path.join(PATH, img_name))
    img.save(os.path.join(OUT_IMG, "train", img_name))

for img_name in val_imgs:
    img = Image.open(os.path.join(PATH, img_name))
    img.save(os.path.join(OUT_IMG, "val", img_name))
```

```python
label_path = "/kaggle/input/vietnamese-license-plate-ocr/labels/crop_labels.csv"
label_dict = {}
with open(label_path, encoding="utf-8") as f:
    reader = csv.reader(f)
    for name, label, _ in reader:
        label_dict[name] = label

def write_txt(imgs, out_file, mode):
    with open(out_file, "w", encoding="utf-8") as f:
        for img in imgs:
            if img in label_dict:
                f.write(f"imgs/{mode}/{img}\t{label_dict[img]}\n")

write_txt(train_imgs, f"{OUT_LABEL}/train.txt", "train")
write_txt(val_imgs,   f"{OUT_LABEL}/val.txt",   "val")
```
